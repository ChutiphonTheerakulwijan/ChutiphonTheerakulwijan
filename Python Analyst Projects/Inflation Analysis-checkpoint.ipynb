## 1. USA Inflation Rate Trend over the Years Analysis

#### Import important Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Import raw inflation data

```python
inflation = pd.read_csv("Inflation Final Pro/inflation interest unemployment.csv")
```

#### Brief look at the Dataset

```python
inflation.head()
```

#### Extract Only USA dataset (Check for completeness of USA data)

```python
inflation[inflation['country'] == "United States"].head()
```

#### Create a new table: Extract only necessary column for analysis, Rename column, Reformat index

```python
usa = inflation[inflation['country'] == "United States"][['year','country','Inflation, consumer prices (annual %)']]
usa.head()
```

```python
usa.rename({'Inflation, consumer prices (annual %)':'Inflation rate'}, axis= 'columns', inplace=True)
usa.head() #Rename Column
```

```python
usa.set_index('year', inplace=True)
usa #Reformat index by use year column
```

#### Prepare Data For Plotting

```python
year = usa.index
year
```

```python
usainflation = usa['Inflation rate']
usainflation.head()
```

#### Plotting

```python
plt.figure(figsize=[10,5])
plt.style.use("fivethirtyeight")

plt.plot(year,usainflation, linewidth=2, color="olive")
plt.title("USA Inflation Rate", fontsize=15)

plt.xlabel("Years", fontsize=10)
plt.ylabel("Inflation rate(%)", fontsize=10)

plt.show()
```

## 2. USA Unemployment Rate Trend over the Years

#### Extract Only USA dataset (Check for Completeness of data)

```python
uu = inflation[inflation['country'] == "United States"][['country', 'year', 'Unemployment, total (% of total labor force) (national estimate)']]
uu.head()
```

#### Create a new table: Extract only necessary column for analysis, Rename column, Reformat index

```python
uu.rename({'Unemployment, total (% of total labor force) (national estimate)': 'Unemployment Rate'}, axis='columns', inplace=True)
uu.head() #Rename Column 
```

```python
uu.set_index("year", inplace=True)
uu #Reformat index by use year column
```

#### Prepare Data for plotting

```python
x = uu.index
y= uu['Unemployment Rate']
```

#### Plotting

```python
plt.figure(figsize=[10,5])

plt.plot(x,y, linewidth=2, color="#CB4335")
plt.title("USA Unemployment Rate", fontsize=15)

plt.xlabel("Years", fontsize=10)
plt.ylabel("Inflation rate(%)", fontsize=10)

plt.show()
```

## 3. Compare USA Inflation Rate and Unemployment Rate Trend over the Years

#### Plotting

```python
plt.figure(figsize=[10,7])

plt.plot(year,usainflation, linewidth=2, color="olive")
plt.plot(x,y, linewidth=2, color="#CB4335")

plt.title("USA Inflation VS. Unemployment Rate Analysis", fontsize=15)
plt.legend(["Inflation Rate","Unemployment Rate"], loc=1, shadow=True, frameon=True, facecolor='white', fontsize=12)

plt.show()
```
